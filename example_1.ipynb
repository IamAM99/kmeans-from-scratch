# Example 1: K-Means Clustering on a Randomly Generated Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_error
from kmeans import KMeans
from sklearn.datasets import make_blobs
```

## Generate A Random Dataset

```python
X = make_blobs(n_samples=1000, n_features=2, centers=3, cluster_std=1.6, random_state=100)[0]

plt.figure()
plt.scatter(X[:, 0], X[:, 1], marker='.')
plt.grid(linestyle='--', linewidth=0.4)
plt.title("Input Dataset")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

## Find The Best Number of Clusters

```python
plot_error(X, list(range(1, 10)))
```

```python
n_clusters = 3
```

## Create A K-Means Model

```python
km = KMeans(n_clusters)
centers, labels = km.fit(X)
```

## Visualize The Clustered Dataset

```python
plt.figure()
for i in np.unique(labels):
    plt.scatter(X[labels == i , 0] , X[labels == i , 1] , label=i, marker='.')
plt.scatter(centers[:, 0], centers[:, 1], marker='x', color='black', s=50, linewidth=2)
plt.legend(['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster Centers'])
plt.grid(linestyle='--', linewidth=0.4)
plt.title("Clustered Dataset")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
