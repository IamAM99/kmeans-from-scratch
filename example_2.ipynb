# Example 2: Image Compression Using K-Means Clustering

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from utils import compress_image
```

```python
# Read the image
img = cv2.imread('./sample_data/bird.jpg')
img = np.flip(img, axis=-1)  # arrange color channels correctly
```

```python
# Compress the image
n_colors = 16
new_img = compress_image(img, n_colors=n_colors)
```

```python
# Show both images
fig, axs = plt.subplots(1, 2, constrained_layout=True)
axs[0].axis('off')
axs[1].axis('off')
axs[0].imshow(img)
axs[0].set_title("Original Image")
axs[1].imshow(new_img)
axs[1].set_title(f"Compressed Image (Colors: {n_colors})")
plt.show()
```

```python
cv2.imwrite("./sample_data/compressed_bird.jpg", np.flip(new_img, axis=-1))
```
